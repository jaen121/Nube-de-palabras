**Nube de puntos ingresando el texto**

```python
# prompt: Word cloud by entering text for words with at least 4 letters

import matplotlib.pyplot as plt
from wordcloud import WordCloud

# Get the text from user input
text = input("Enter your text: ")

# Create a dictionary of word frequencies
word_counts = {}
for word in text.split():
    if len(word) >= 4:
        word_counts[word] = word_counts.get(word, 0) + 1

# Create the word cloud
wordcloud = WordCloud(width=800, height=600, background_color="white").generate_from_frequencies(word_counts)

# Display the word cloud
plt.figure(figsize=(10, 7))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```

```python
# Instalar las bibliotecas necesarias
#!pip install pdfminer.six
#!pip install wordcloud

from pdfminer.high_level import extract_text
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from google.colab import drive

# Montar Google Drive
drive.mount('/content/drive')

# Ruta del archivo PDF en Google Drive
filename = "/content/drive/MyDrive/Nube de palabras/Nube de palabras.pdf"

# Extraer texto del PDF
text = extract_text(filename)

# Filtrar palabras de más de 4 letras
filtered_text = ' '.join([word for word in text.split() if len(word) > 6])

# Crear la nube de palabras
wordcloud = WordCloud().generate(filtered_text)

# Mostrar la nube de palabras
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
